### Drawing Vectors & Matrices

#### 2D vectors using matplotlib

[`matplotlib.pyplot.quiver(*args, data=None, **kw): quiver([X, Y], U, V, [C], **kw)`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.quiver.html)

Where X, Y define the arrow locations, U, V define the arrow directions, and C optionally sets the color.

````python
# Define vectors to plot: V1 and V2
V1 = np.array([2,0])
V2 = np.array([0,3])

print('Vector plot V1 and V2: \n {} + {}'.format(V1, V2))

origin = [0],[0] #=> set origin

plt.quiver(*origin, *V1, color=['r'], units='xy', scale=1) # plot the quiver
# OR
# plt.quiver(*origin, V1[0], V1[1], color=['r'], units='xy', scale=1)
plt.quiver(*origin, *V2, color=['g'], units='xy', scale=1)

plt.grid(linestyle=':', linewidth=.1)

# Axis limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Add a legend
plt.legend(['V'+str(i+1) for i in range(2)]) #<-- give a legend

plt.show()
````

````python
# Vector addition example
V_add = V1+V2

print('Vector sum V1 + V2 = V_add: \n {} + {} => {}'.format(V1, V2, V_add))

plt.quiver(*origin, *V1, color=['r'], units='xy', scale=1)
plt.quiver(*origin, *V2, color=['g'], units='xy', scale=1)
plt.quiver(*origin, *V_add, color=['b'], units='xy', scale=1)

plt.grid(linestyle=':', linewidth=.1)

# Axis limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Add a legend
plt.legend(['V2', 'V2', 'V_add']) #<-- give a legend

plt.show()
````

````python
# Vector subtraction example
V_sub = V1-V2

print('Vector subtraction V1 - V2 = V_sub: \n {} - {} => {}'.format(V1, V2, V_sub))

plt.quiver(*origin, *V1, color=['r'], units='xy', scale=1)
plt.quiver(*origin, *V2, color=['g'], units='xy', scale=1)
plt.quiver(*origin, *V_sub, color=['b'], units='xy', scale=1)

plt.grid(linestyle=':', linewidth=.1)

# Axis limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Add a legend
plt.legend(['V1','V2','V_sub']) #<-- give a legend

plt.show()
````

#### 3D vectors using matplotlib

[`Axes3D.quiver(*args, **kwargs): quiver(X, Y, Z, U, V, W, **kwargs)`](https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html)

`X, Y, Z`: The x, y and z coordinates of the arrow locations (default is tail of arrow; see pivot kwarg)

`U, V, W`: The x, y and z components of the arrow vectors

````python
# x, y, z vector start locations
origin = np.array([0,0,0])

# vectors
V1 = np.array([1,3,2])
V2 = np.array([-1,1,3])

print('Arrow locations and vectors')
print('Origin (x, y, z): {}'.format(origin))
print('V1: {}\nV2: {}'.format(V1, V2))

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.quiver(*origin, *V1, color='red')
ax.quiver(*origin, *V2, color='green')

ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

plt.show()
````

Plotting multiple 3D vectors at the same time

### Cross Products

The cross product `a  b` is defined as a vector `c` that is perpendicular (orthogonal) to both `a` and `b`. It will also has a direction given by the right-hand rule and a magnitude <span style="color:blue;">**equal to the area of the parallelogram**</span> that the vectors span.

![matrix_network](img/vector_cross_product.png)

**Starting with a 2D example to make visualization easier**

In Python, we create the above image by defining a and b, then c as the cross product between the two.

````python
# Define vector a and vector b
a = np.array([[2,3]])
b = np.array([[1,-3]])

# Compute their cross product and assign it to c
c = np.cross(a,b)
````

````python
# Print out the shape (ie dimensions) of a, b, and c
print('Dimension of vector a is {} (pronounced {} by {})'.format(a.shape, a.shape[0], a.shape[1]))
print('Dimension of vector b is {} (pronounced {} by {})'.format(b.shape, b.shape[0], b.shape[1]))
print('Dimension of vector c is {} (pronounced {} by {})'.format(c.shape, c.shape[0], c.shape[0]))
````

### Dot Products

The dot product (or scalar product) takes to matricies and gives a single scalar number. You can take the dot product of any two matricies that share the inner dimension.

To multiply two vectors it makes sense to multiply their lengths together but only when they point in the same direction, so we only use the component that is pointed in the same direction (ie. cos(theta))

a  b = |a|  |b|  cos()

a  b = |a|  |b|  cos(90)
a  b = |a|  |b|  0
a  b = 0
or we can calculate it this way:

a  b = ax  bx + ay  by

a  b = -12  12 + 16  9
a  b = -144 + 144
a  b = 0

Algebraically the dot product is the sum of products:

$$
{\displaystyle {\begin{bmatrix}\color {red}1&\color {red}3&\color {red}-5\end{bmatrix}}{\begin{bmatrix}\color {blue}4\\\color {blue}-2\\\color {blue}-1\end{bmatrix}}=\color {purple}3}.
$$

Geometrically, the dot product is the magnitude of the vectors cos the angle between them or ${\displaystyle \mathbf {a} \cdot \mathbf {b} =\|\mathbf {a} \|\ \|\mathbf {b} \|\cos \theta ,}$. Geometrically it looks like this:

<img src="img/scalar_projection.png" alt="scalar projection" width="200"/>

We will use these vectors in the examples that follow

````python
d = np.array([[2,3]]); print(d.shape)
e = np.array([[1],[5]]); print(e.shape)
f = np.array([[1],[3],[-5]]); print(f.shape)
g = np.array([[4,-2,-1]]); print(g.shape)
````

You see we can get the dot product of the `(1x2)` and `(2x1)` matrix because they share the inner dimension `2`

````python
d_dot_e = np.dot(d,e); print('d dot e: {}'.format(d_dot_e))
f_dot_g = np.dot(f,g); print('g dot g: {}'.format(f_dot_g))
e_dot_g = np.dot(e,g); print('e dot g: {}'.format(e_dot_g))
````
